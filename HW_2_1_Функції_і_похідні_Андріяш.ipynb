 1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)
```

```python

# Створення графіку
plt.figure(figsize=(12, 6))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Позначення точки та дотичної лінії
x_point = 5
f_point = 10 * x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red')  # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 5')  # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("Години навчання (x)")
plt.ylabel("Оцінка (y)")
plt.legend()
plt.grid(True)

plt.show()
```

Кожна додаткова година навчання додає +10 до оцінки. Графік прямий, що показує стабільне зростання.

 2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



f(x) = sin(x):

```python
# Визначаємо символічну змінну
x = sp.Symbol('x')

# Функція f(x) = sin(x)
f = sp.sin(x)
# Обчислюємо похідну
f_prime = sp.diff(f, x)

# Перетворення у числові функції
f_lambdified = sp.lambdify(x, f, 'numpy')
f_prime_lambdified = sp.lambdify(x, f_prime, 'numpy')

# Діапазон значень для графіка
x_vals = np.linspace(0, 2 * np.pi, 400)

# Обчислення значень функції та похідної
f_vals = f_lambdified(x_vals)
f_prime_vals = f_prime_lambdified(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = cos(x)", color='green')
plt.title('Функція sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

  f(x)=ln(x)

```python
# Визначаємо символічну змінну
x = sp.Symbol('x')

# Функція f(x) = ln(x)
f = sp.ln(x)
# Обчислюємо похідну
f_prime = sp.diff(f, x)

# Перетворення у числові функції
f_lambdified = sp.lambdify(x, f, 'numpy')
f_prime_lambdified = sp.lambdify(x, f_prime, 'numpy')

# Діапазон значень для графіка
x_vals = np.linspace(0.1, 10, 400)

# Обчислення значень функції та похідної
f_vals = f_lambdified(x_vals)
f_prime_vals = f_prime_lambdified(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 1/x", color='green')
plt.title('Функція ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.Symbol('x')

# Функція f(x) = x^2 - 5x + 1
f = x**2 - 5*x + 1
# Обчислюємо похідну
f_prime = sp.diff(f, x)

# Перетворення у числові функції
f_lambdified = sp.lambdify(x, f, 'numpy')
f_prime_lambdified = sp.lambdify(x, f_prime, 'numpy')

# Діапазон значень для графіка
x_vals = np.linspace(0, 100, 200)

# Обчислення значень функції та похідної
f_vals = f_lambdified(x_vals)
f_prime_vals = f_prime_lambdified(x_vals)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_vals, f_prime_vals, label="f'(x) = 2x - 5", color='green')
plt.title('Загальні витрати компанії та їх похідна')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Витрати (y)')
plt.legend()
plt.grid(True)
plt.show()
```
